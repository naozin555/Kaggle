# 機械学習アルゴリズムを変更
## ランダムフォレスト、LightGBM（勾配ブースティング）を試す

### 以前のコード

```python
import numpy as np
import pandas as pd


train = pd.read_csv('../input/titanic/train.csv')
test = pd.read_csv('../input/titanic/test.csv')
gender_submission = pd.read_csv('../input/titanic/gender_submission.csv')

data = pd.concat([train, test], sort=False)

data['Sex'].replace(['male', 'female'], [0, 1], inplace=True)
data['Embarked'].fillna(('S'), inplace=True)
data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)
data['Fare'].fillna(np.mean(data['Fare']), inplace=True)
data['Age'].fillna(data['Age'].median(), inplace=True)
data['FamilySize'] = data['Parch'] + data['SibSp'] + 1
data['IsAlone'] = 0
data.loc[data['FamilySize'] == 1, 'IsAlone'] = 1
```

```python
data.head()
```

```python
delete_columns = ['Name', 'PassengerId', 'Ticket', 'Cabin']
data.drop(delete_columns, axis=1, inplace=True)

train = data[:len(train)]
test = data[len(train):]

y_train = train['Survived']
X_train = train.drop('Survived', axis=1)
X_test = test.drop('Survived', axis=1)
```

```python
X_train.head()
```

# ランダムフォレスト
- 訓練データからn個のブートストラップデータを取り出す
- これらを使ってn個の木を生成する
  - この時p個の特徴量(Pclass, Name, Age, ..., IsAlone)からm個だけをランダムに選ぶ
- 回帰の場合は平均を、分類の場合は多数決をとって最終的な予測値とする

```python
# 分類の場合は、RandomForestClassifierクラス
# 回帰の場合は、RandomForestRegressorクラス
from sklearn.ensemble import RandomForestClassifier

# n_estimators：tree数
# max_depth：木の深さ（経験則として「分類」：n=1、「回帰」：n=5にすると良い結果が得られるらしい。）
# random_state：乱数のseed
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
sub = pd.read_csv('../input/titanic/gender_submission.csv')
sub['Survived'] = list(map(int, y_pred))
sub.to_csv('submission_randomforest.csv', index=False)
sub.head()
```

# LightGBM(勾配ブースティング)
- ベース学習器を逐次的に訓練する

```python
from sklearn.model_selection import train_test_split

# 学習用データと検証用データの分割
# test_size：検証用データセットに用いる割合
# random_state：乱数のseed
# stratify：分割後の学習用データセットと検証用データセットの割合
X_train, X_valid, y_train, y_valid = \
    train_test_split(X_train, y_train, test_size=0.3,
                                 random_state=0, stratify=y_train)

# 学習用データセットと検証用データセットの作成
import lightgbm as lgb
categorical_features = ['Embarked', 'Pclass', 'Sex']
lgb_train = lgb.Dataset(X_train, y_train,
                                         categorical_feature=categorical_features)
lgb_eval = lgb.Dataset(X_valid, y_valid, reference=lgb_train,
                                         categorical_feature=categorical_features)
params = {
    'objective': 'binary'
}

# 学習と予測
# num_boost_round：決定木の数
# early_stopping_rounds:何回連続で予測の精度が上がらなかったら終了
# 今回は38回以降の計算で10回連続精度が向上しなかったため、打ち切り
model = lgb.train(params, lgb_train,
                               valid_sets=[lgb_train, lgb_eval],
                               verbose_eval=10,
                               num_boost_round=1000,
                               early_stopping_rounds=10)

y_pred = model.predict(X_test, num_iteration=model.best_iteration)
```

```python
y_pred[:10]
```

```python
# 予測結果が連続値なので離散値に変換
y_pred = (y_pred > 0.5).astype(int)
y_pred[:10]
```

```python
# 提出用ファイル作成
sub['Survived'] = y_pred
sub.to_csv('submission_lightgbm.csv', index=False)

sub.head()
```
